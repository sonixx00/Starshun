```python
import numpy as np

# Таблиця значень функції
x_values = np.array([4.5, 5.0, 5.5, 6.0, 6.5, 7.0])
y_values = np.array([8.442, 8.482, 8.862, 9.701, 11.132, 13.302])

# Крок таблиці
h = 0.5

# Функція для обчислення кінцевих різниць
def finite_differences(y):
    n = len(y)
    delta_y = np.zeros((n, n))
    delta_y[:, 0] = y
    for j in range(1, n):
        for i in range(n - j):
            delta_y[i, j] = delta_y[i + 1, j - 1] - delta_y[i, j - 1]
    return delta_y

# Отримуємо таблицю кінцевих різниць
delta_table = finite_differences(y_values)

# Індекс для x = 5.5 у масиві x_values це 2
idx = 2

# Значення кінцевих різниць для точки x=5.5
dy2 = delta_table[idx, 1]   # 0.839
d2y2 = delta_table[idx, 2]  # 0.592
d3y2 = delta_table[idx, 3]  # 0.147

# Обчислення першої похідної (формула для вузла, q=0)[cite: 3]
y_prime = (1/h) * (dy2 - (d2y2 / 2) + (d3y2 / 3))

# Обчислення другої похідної (формула для вузла, q=0)[cite: 3]
y_double_prime = (1/(h**2)) * (d2y2 - d3y2)

# Вивід результатів
print("Таблиця кінцевих різниць:")
np.set_printoptions(precision=4, suppress=True)
print(delta_table)

print(f"\nРезультати для x = {x_values[idx]}:")
print(f"Перша похідна y' ≈ {round(y_prime, 3)}")
print(f"Друга похідна y'' ≈ {round(y_double_prime, 3)}")
```
