```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Вихідна функція для варіанта 18
def func(x):
    return np.sin(2 * x) + x

# 2. Формування вхідних даних (таблиця xi та yi)
x = np.array([i * 0.1 for i in range(1, 11)]) # [0.1, 0.2, ..., 1.0]
y = func(x)

print(f"x = {x}")
print(f"y = {np.round(y, 4)}")

# --- ГРАФІК 1: НАБЛИЖЕННЯ ПРЯМОЮ (Лінійна регресія) ---
# Знаходження коефіцієнтів a та b для y = ax + b
coef1 = np.polyfit(x, y, 1)
poly1 = np.poly1d(coef1)

plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='red', label='Точки даних')
plt.plot(x, poly1(x), color='blue', label=f'Лінія: y = {coef1[0]:.2f}x + {coef1[1]:.2f}')
plt.title('МНК: Наближення прямою (Варіант 18)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# --- ГРАФІК 2: НАБЛИЖЕННЯ ПАРАБОЛОЮ (Поліном 2-го порядку) ---
# Знаходження коефіцієнтів a, b, c для y = ax^2 + bx + c
coef2 = np.polyfit(x, y, 2)
poly2 = np.poly1d(coef2)

# Точки для побудови плавної лінії параболи
x_smooth = np.linspace(min(x), max(x), 100)

plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='red', label='Точки даних')
plt.plot(x_smooth, poly2(x_smooth), color='green',
         label=f'Парабола: y = {coef2[0]:.2f}x² + {coef2[1]:.2f}x + {coef2[2]:.2f}')
plt.title('МНК: Наближення параболою (Варіант 18)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# --- ГРАФІК 3: ПОРІВНЯННЯ ОБОХ МОДЕЛЕЙ НА ОДНОМУ ПОЛІ ---
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'ro', label='Точки f(x)')
plt.plot(x_smooth, poly1(x_smooth), '--', color='blue', label='Лінійне наближення')
plt.plot(x_smooth, poly2(x_smooth), color='green', label='Параболічне наближення')
plt.title('Порівняння точності наближень (Варіант 18)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
