```python
import numpy as np
from scipy import integrate

a = 1.2
b = 2.0
n = 10
h = (b - a) / n

def f1(x):
    return 1 / np.sqrt(0.5 * x + 1)

# Метод лівих прямокутників
def left_rec(f, a, b, n, h):
    sum_val = 0
    for i in range(0, n):
        sum_val += f(a + i * h)
    return sum_val * h

# Метод правих прямокутників
def right_rec(f, a, b, n, h):
    sum_val = 0
    for i in range(1, n + 1):
        sum_val += f(a + i * h)
    return sum_val * h

# Метод середніх прямокутників
def aver_rec(f, a, b, n, h):
    sum_val = 0
    for i in range(0, n):
        # Використовуємо значення в середині відрізка
        sum_val += f(a + (i + 0.5) * h)
    return sum_val * h

print("--- METOD PRYAMOKUTNYKIV ---")
print(f"Left rectangle: {round(left_rec(f1, a, b, n, h), 5)}")
print(f"Right rectangle: {round(right_rec(f1, a, b, n, h), 5)}")
print(f"Average rectangle: {round(aver_rec(f1, a, b, n, h), 5)}")

# Перевірка
v, err = integrate.quad(f1, a, b)
print(f"Check with SciPy: {round(v, 5)}")
```

```python
def f2(x):
    return np.log10(x**2 + 3) / (2 * x)

a2 = 1.2
b2 = 2.0
n2 = 8
h2 = (b2 - a2) / n2

def simpson_rule(f, a, b, n, h):
    integr = f(a) + f(b)
    for i in range(1, n):
        x = a + i * h
        if i % 2 == 0:
            integr += 2 * f(x) # Парні індекси
        else:
            integr += 4 * f(x) # Непарні індекси
    return integr * (h / 3)

print("\n--- METOD SIMPSONA ---")
res_simpson = simpson_rule(f2, a2, b2, n2, h2)
print(f"Simpson method result: {round(res_simpson, 5)}")

v2, err2 = integrate.quad(f2, a2, b2)
print(f"Check with SciPy: {round(v2, 5)}")
```

```python
def f3(x):
    return 1 / np.sqrt(x**2 + 0.6)

a3 = 1.2
b3 = 2.6
n3 = 20
h3 = (b3 - a3) / n3

def trapezoidal_rule(f, a, b, n, h):
    # Сума півсуми крайніх значень та всіх внутрішніх
    sum_val = 0.5 * (f(a) + f(b))
    for i in range(1, n):
        sum_val += f(a + i * h)
    return sum_val * h

print("\n--- METOD TRAPETZIY ---")
res_trap = trapezoidal_rule(f3, a3, b3, n3, h3)
print(f"Trapezoidal method result: {round(res_trap, 5)}")

v3, err3 = integrate.quad(f3, a3, b3)
print(f"Check with SciPy: {round(v3, 5)}")
```
